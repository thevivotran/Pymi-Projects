## Sin, Cos, Tan Graph (-2ùúã -> 2ùúã)

```python
x = np.linspace(-2*np.pi, 2*np.pi)
y = np.linspace(-1, 1)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, [0 for i in np.linspace(-7, 7)])
plt.plot([0 for i in np.linspace(-1, 1)], y)
plt.legend(['y=sin(x)', 'y=0', 'x=0'])
plt.title('Sin Graph')
```

```python
plt.plot(x, np.cos(x))
plt.plot(x, [0 for i in np.linspace(-7, 7)])
plt.plot([0 for i in np.linspace(-1, 1)], y)
plt.legend(['y=cos(x)', 'y=0', 'x=0'])
plt.title('Cos Graph')
```

```python
plt.plot(x, np.tan(x))
plt.plot(x, [0 for i in x])
plt.plot([0 for i in y], y)
plt.legend(['y=tan(x)', 'y=0', 'x=0'])
plt.title('Tan Graph')
```
