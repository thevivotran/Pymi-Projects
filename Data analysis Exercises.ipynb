## Sin, Cos, Tan Graph (-2ùúã -> 2ùúã)

```python
x = np.linspace(-2*np.pi, 2*np.pi)
y = np.linspace(-1, 1)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, [0 for i in np.linspace(-7, 7)])
plt.plot([0 for i in np.linspace(-1, 1)], y)
plt.legend(['y=sin(x)', 'y=0', 'x=0'])
plt.title('Sin Graph')
```

```python
plt.plot(x, np.cos(x))
plt.plot(x, [0 for i in np.linspace(-7, 7)])
plt.plot([0 for i in np.linspace(-1, 1)], y)
plt.legend(['y=cos(x)', 'y=0', 'x=0'])
plt.title('Cos Graph')
```

```python
plt.plot(x, np.tan(x))
plt.plot(x, [0 for i in x])
plt.plot([0 for i in y], y)
plt.legend(['y=tan(x)', 'y=0', 'x=0'])
plt.title('Tan Graph')
```

## Case Study 2

### Find the day that has the highest price of BTC

```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('localbtcVND.csv', header=None)
```

```python
df = df.rename(columns={0:'Date', 1:'Unknown', 2:'Value'})
```

```python
df
```

```python
def unix_to_date(unix):
    rs = datetime.fromtimestamp(unix)
    return '{}-{}-{}'.format(str(rs.day).zfill(2), str(rs.month).zfill(2), str(rs.year))
```

```python
df['Date'] = df['Date'].apply(unix_to_date)
```

```python
df = df.set_index('Date')
```

```python
df
```

```python
result = df.sort_values(by='Value', ascending=False)
```
